# Average True Range

The average true range is a historical volatility indicator taking the moving average of the true range of an asset. The true range is defined as the difference between the high and the low price of a an asset during a given period. This indicator is important because it gives the analyst a better idea of how the volatility of an asset has changed throughout time. 

The average true range can be calculated easily using Python. This notebook will take to the process to calculate the true range of any stock and to quickly analyze the historical trend of the true range.

Let's first import all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
from pandas_datareader import data as wb
```

```python
#Dictionaries to customize the font of labels and titles of graphs
label_dict = {'fontsize': 14,
            'fontweight' : 5,
            'verticalalignment': 'baseline',
            'horizontalalignment': 'center'
             }
title_dict = {'fontsize': 20,
           'fontweight' : 5,
           'verticalalignment': 'baseline',
           'horizontalalignment': 'center'
             }
```

The analysis presented in this notebook is for the S&P500 over a period of 38 years, from 1983 to 2021. This period of time will give us an idea of how the volatility has changed in the market in the last four decades.

```python
stock = '^GSPC'
start = '1983-01-01'
```

We will need to scrap the price data from Yahoo Finance. The module <code>pandas_datareader</code> can retreive this information easily. We can save this data in a DataFrame called <code>data</code>. As noticed below, The DataFrame contains information about High, Low, Open, Close, Volume and Adjusted Close price on the S&P500.

```python
data = pd.DataFrame()
data = wb.DataReader(stock, data_source='yahoo', start = start)

data
```

Let's add one column that will equal the difference between the High and Low price. This column is the True Range of the S&P500 on a daily basis. The True Range is the extreme change that happened during one trading day. The limitation of the True Range is that it is expressed in absolute values which makes comparison difficult. One way to overcome this limitation is by calculating the True Range as a percentage.

```python
data['True Range'] = data['High'] - data['Low']
data
```

As mentioned before, a better way to look the True Range is in percentage basis. To achive this, we can take the True Range column and calculate the two days moving average of the True Range and then divide it over the Open price. This is the Average True Range over one day. Just noticed that with this transformation, we are losing one data point. If we increase the number of days the moving average is calculated, we will lose more data points.

```python
data['ATR 1D'] = data['True Range'].rolling(2).mean()/data['Open']

data
```

Let's have an idea of the historical Average True Range in the market of the last 40 years. 

```python
fig,axs = plt.subplots(2,1,figsize=(15,10))

fig.subplots_adjust(hspace=0)

axs[0].plot(data['Adj Close'],label='S&P500 Price')
axs[0].set_xlim(xmin=data['Adj Close'].index[0],xmax=data['Adj Close'].index[-1])
axs[0].set_ylim(ymin=0)
axs[0].legend(labelspacing=2,borderpad=1.25,fontsize=10,loc=9)
axs[0].set_ylabel('Price ($)',fontdict=label_dict,labelpad=20)

axs[1].plot(data['ATR 1D'],label='Average True Range',c='g')
axs[1].set_xlim(xmin=data['ATR 1D'].index[0],xmax=data['ATR 1D'].index[-1])
axs[1].set_ylim(ymin=0, ymax=0.22)
axs[1].legend(labelspacing=2,borderpad=1.25,fontsize=10)
axs[1].set_xlabel('Time', fontdict=label_dict,labelpad=20)
axs[1].set_ylabel('Implied Volatility (%)', fontdict = label_dict,labelpad=10)

axs[0].axvspan(data['ATR 1D'].index[1130], data['ATR 1D'].index[1280], color='red', alpha=0.20)
axs[1].axvspan(data['ATR 1D'].index[1130], data['ATR 1D'].index[1280], color='red', alpha=0.20)

axs[0].axvspan(data['ATR 1D'].index[6400], data['ATR 1D'].index[6630], color='red', alpha=0.20)
axs[1].axvspan(data['ATR 1D'].index[6400], data['ATR 1D'].index[6630], color='red', alpha=0.20)

axs[0].axvspan(data['ATR 1D'].index[9330], data['ATR 1D'].index[9400], color='red', alpha=0.20)
axs[1].axvspan(data['ATR 1D'].index[9330], data['ATR 1D'].index[9400], color='red', alpha=0.20)

axs[0].set_title('SPDR S&P500 - Market Price vs Average True Range',fontdict=title_dict,pad=20)


plt.show()
```

The graph above shows that the average volatility in the market has decreased considerably over the time. Noticed that the spike in volatility sorrounding the crash of 1987 is considerably higher of the two most recent market crashes, the 2008 financial crisis and the COVID-19 crisis. This is a helpful insight to understand potential opportunities in the market and the time horizon of this opportunities.

Let's take a look to the Average True Range over different period of times. In the next chart, we will calculate the average historical volatility over one week, one month, one year, 3 years, 5 years, 10 years, 20 years, and 30 years.

```python
columns = {'1 month':20,'6 months':175,'1 year':250, '3 years':750, '5 years':1250,
           '10 years':2500,'20 years':5000,'30 years':7500,'38 years':9500
          }
index = ['Average %']

df = pd.DataFrame(index=index)

for i,j in columns.items():
    df[i] = round(data['ATR 1D'].iloc[-j:].mean()*100,3)
    
df
```

As noted in the table above, daily volatility increased during the 3 year period and considerable exploded one year ago. This was during the COVID-19 crisis. Volatility has decreased during this last year, getting back to the long-term trend. In the long-term, daily volatility decreased consistently over 30, 20, 10 and 5 years; this last one being the lowest of the analysis. 

__Conclusion:__

The Average True Range helps to understand volatility changes in assets. Knowing how volatility has changed over time is important because gives the analyst an understanding on where opportunities are in the market. When daily volatility increase, trader can make money over short periods of time; however, when daily volatility is low, chances to make money over short periods of time is difficult.
